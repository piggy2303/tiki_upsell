```python
from datetime import datetime
from elasticsearch import Elasticsearch
from elasticsearch.helpers import bulk
import time

es = Elasticsearch(port=9201)
es.indices.create(index='tiki', ignore=400)

def bulk_sync(items):
    actions = [
        {
            "_index": "tiki",
            "_source": item
        } for item in items
    ]
    bulk(es, actions)

time_start = time.time()
bulk_sync(myArray)
print(time.time() - time_start)
```

```python
list_file = os.listdir("./data_graph/")
list_file

done_data_graph = open("done_data_graph.txt","a")

for file_npy in list_file:
    print("doing",file_npy)
    myArray = np.load("./data_graph/"+file_npy,allow_pickle=True)
    bulk_sync(myArray)
    
    done_data_graph.write(file_npy+"\n")
done_data_graph.close()
```
